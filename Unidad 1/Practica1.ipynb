```python
semilla=7458
n=4
iteraciones=100
resultados=[]

for _ in range(iteraciones):
    cuadrado=str(semilla**2).zfill(2*n)
    inicio=(len(cuadrado)-n)//2
    semilla=int(cuadrado[inicio:inicio+n])
    resultados.append(semilla)

for i, numero in enumerate(resultados, 1):
    print(f"Número aleatorio {i}: {numero}")
```

```python
import random
semilla1 = random.randint(1000, 9999)
semilla2 = random.randint(1000, 9999)
n = 4
iteraciones = 100
resultados = []
for _ in range(iteraciones):
  producto = str(semilla1 * semilla2).zfill(2*n)
  inicio = (len(producto) - n) // 2
  nuevo_numero = int(producto[inicio:inicio+n])
  resultados.append(nuevo_numero)
  semilla1, semilla2 = semilla2, nuevo_numero

print(semilla1, "y ",semilla2)
for i, numero in enumerate(resultados, 1):
    print(f"Número aleatorio {i}: {numero}")
```
