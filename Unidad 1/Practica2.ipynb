```python
import numpy as np
from scipy.stats import kstest

random_numbers = np.random.uniform(0, 1, 100)
n=0
print("Números aleatorios generados:")
for num in random_numbers:
  n+=1
  print("numero ",n,": ",num)

print("\nResultados de la prueba Kolmogorov-Smirnov para cada número:")
n=0
for num in random_numbers:
    ks_statistic, p_value = kstest([num], 'uniform')
    n+=1
    print(f"Número {n}: {num}, Estadístico KS: {ks_statistic}, Valor p: {p_value}")

ks_statistic, p_value = kstest(random_numbers, 'uniform')

print("\nResultados generales de la prueba Kolmogorov-Smirnov:")
print(f"Estadístico de Kolmogorov-Smirnov: {ks_statistic}")
print(f"Valor p: {p_value}")

alpha = 0.05
if p_value > alpha:
    print("No se rechaza la hipótesis nula: los datos siguen una distribución uniforme.")
else:
    print("Se rechaza la hipótesis nula: los datos no siguen una distribución uniforme.")
```
