```python
import random

m = 2**31 - 1
a = 1664525
c = 1013904223
X0 = random.randint(1, m-1)


Xn = X0
random_numbers = []
for _ in range(100):
    Xn = (a * Xn + c) % m
    random_numbers.append(Xn / m)

for i, num in enumerate(random_numbers[:100], 1):
    print(f"Número {i}: {num}")
```
