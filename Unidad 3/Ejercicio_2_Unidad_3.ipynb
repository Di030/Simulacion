```python
import numpy as np
import matplotlib.pyplot as plt

def simulacion_llegadas_vehiculos():
    tasa_llegadas = 1/15
    tiempo_total = 12 * 60
    np.random.seed(42)

    tiempos_entre_llegadas = []
    tiempo_acumulado = 0
    while tiempo_acumulado < tiempo_total:
        tiempo = np.random.exponential(scale=1/tasa_llegadas)
        tiempo_acumulado += tiempo
        if tiempo_acumulado < tiempo_total:
            tiempos_entre_llegadas.append(tiempo)

    tiempos_entre_llegadas = np.array(tiempos_entre_llegadas)

    print("\n--- Resultados de la simulación ---")
    print(f"Total de vehículos llegados: {len(tiempos_entre_llegadas)}")
    print(f"Tiempo promedio entre llegadas: {np.mean(tiempos_entre_llegadas):.2f} minutos")
    print(f"Desviación estándar: {np.std(tiempos_entre_llegadas):.2f} minutos")
    print(f"Tiempo mínimo entre llegadas: {np.min(tiempos_entre_llegadas):.2f} minutos")
    print(f"Tiempo máximo entre llegadas: {np.max(tiempos_entre_llegadas):.2f} minutos")

    plt.figure(figsize=(10, 6))
    plt.hist(tiempos_entre_llegadas, bins=30, edgecolor='black', alpha=0.7)
    plt.title('Distribución de tiempos entre llegadas de vehículos eléctricos')
    plt.xlabel('Tiempo entre llegadas (minutos)')
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.grid(True)
    plt.show()

    return tiempos_entre_llegadas

tiempos = simulacion_llegadas_vehiculos()
```

La distribución exponencial demostró ser la elección ideal para modelar los tiempos entre llegadas de vehículos eléctricos a la estación de carga, ya que:

Modela procesos sin memoria: La propiedad de falta de memoria de la exponencial significa que la probabilidad de una llegada no depende del tiempo transcurrido desde el último evento, lo cual es realista para llegadas aleatorias e independientes.

Captura la naturaleza aleatoria de las llegadas: A diferencia de distribuciones simétricas (como la normal), la exponencial refleja adecuadamente que la mayoría de los intervalos son cortos, pero ocasionalmente pueden ocurrir tiempos más largos.

Proporciona resultados prácticos para la gestión de recursos: Al simular los tiempos entre llegadas durante 12 horas, se puede estimar la demanda máxima y optimizar la cantidad de puntos de carga necesarios.

Beneficios del Método Exponencial
Simplicidad y eficiencia computacional

Solo requiere un parámetro (tasa de llegadas λ = 1/15 min⁻¹).

Fácil de implementar en Python con numpy.random.exponential.

Útil para análisis de colas y capacidad

Permite predecir cuántos vehículos llegarán en intervalos específicos.

Ayuda a dimensionar la infraestructura (puntos de carga) para evitar saturación.

Identificación de picos de demanda

Detecta momentos con llegadas muy cercanas (congestión) o muy espaciadas (tiempos muertos).

Permite ajustar horarios de personal o priorizar carga rápida en horas críticas.

Escalabilidad

El modelo puede extenderse a diferentes tasas de llegada (ej: mañana vs. tarde).

Se puede combinar con simulaciones de tiempo de carga para un análisis más completo.
