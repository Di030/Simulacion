```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def simulacion_pedidos_semana():
    n_pedidos = 20
    p_exito = 0.70
    dias_semana = 7
    np.random.seed(42)

    pedidos_satisfechos = binom.rvs(n=n_pedidos, p=p_exito, size=dias_semana)
    print("\n--- Resultados de la simulación ---")
    print(f"Pedidos satisfechos por día: {pedidos_satisfechos}")
    print(f"Promedio diario de pedidos satisfechos: {np.mean(pedidos_satisfechos):.2f}")
    print(f"Máximo pedidos satisfechos en un día: {np.max(pedidos_satisfechos)}")
    print(f"Mínimo pedidos satisfechos en un día: {np.min(pedidos_satisfechos)}")
    print(f"Total de pedidos satisfechos en la semana: {np.sum(pedidos_satisfechos)}")

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.bar(range(1, dias_semana+1), pedidos_satisfechos, color='skyblue')
    plt.title('Pedidos satisfechos por día')
    plt.xlabel('Día de la semana')
    plt.ylabel('Pedidos satisfechos')
    plt.xticks(range(1, dias_semana+1))
    plt.grid(axis='y')

    plt.show()

    return pedidos_satisfechos

resultados = simulacion_pedidos_semana()
```

El método binomial demostró ser ideal para modelar situaciones de éxito/fracaso discretas, como en el caso del distribuidor electrónico. A través de esta simulación:

Precisión en la planificación: Permitió proyectar con exactitud cuántos pedidos podrían satisfacerse diariamente, considerando la variabilidad natural del proceso.

Gestión de inventario óptima: Al conocer la distribución probable de pedidos exitosos, el distribuidor puede ajustar sus niveles de stock para minimizar tanto excesos como faltantes.

Identificación de riesgos: La simulación reveló el rango esperado de pedidos satisfechos, permitiendo prepararse para escenarios extremos (días con muchos pedidos no satisfechos).

Cálculo de recursos necesarios: Al extender el análisis a una semana completa, se obtiene una visión agregada útil para la planificación semanal de personal y logística.

Beneficios:

Modelado de eventos discretos: Perfecto para contar éxitos/fracasos (pedidos satisfechos/no satisfechos)

Simpleza interpretativa: Solo requiere dos parámetros (n, p) fáciles de entender

Eficiencia computacional: Cálculos rápidos incluso para grandes volúmenes

Aplicación directa: Ideal para problemas de gestión de inventarios y control de calidad
